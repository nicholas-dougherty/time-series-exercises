````python
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 7))

sns.scatterplot(data=df, y ='mpg', x= 'displacement', ax=ax1)
sns.scatterplot(data=df, y='mpg', x='weight', ax=ax2)
                
                
ax1.set(title='Displacement Vs Mpg')
ax2.set(title='Weight vs MPG')
fig.suptitle('Cars MPG Visualizations')
````
